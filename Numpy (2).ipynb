```python
import numpy as np
```

```python
#ARRAY OF ONES
ones = np.ones((4,3,4),dtype=int)
print(ones)
```

```python
#ARRAY OF ZEROS
zeros = np.zeros((5,3,4),dtype=np.int)
print(zeros)
```

```python
#ARRAY WITH RANDOM VALUES
s = np.random.random((2,2))
print(s)
```

```python
#EMPTY ARRAY
empty_array = np.empty((3,2))
print(empty_array)
```

```python
#FULL ARRAY
full_array = np.full((3,2),7)
print(full_array)
```

```python
#ARRAY OF EVENLY SPACED VALUES
evenspaced_array = np.arange(10,25,5)
print(evenspaced_array)
```
