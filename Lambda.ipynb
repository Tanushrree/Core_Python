```python
s = [1,3,5,7,9]
s_new = list(map(lambda s:s*7,s))
print(s_new)
```

```python
name = ['prita','tanu','ammu']
name_new = list(filter(lambda name:(name=='tanu'),name))
print(name_new)
```

```python
name_new = list(map(lambda name : name.upper(),name))
print(name_new)
```
